```python
from pathlib import Path
import json, pprint
from IPython.display import Image, display

BASE = Path(".").resolve().parents[1]  # repo root
print("Repo root:", BASE)

# Collect and present metrics and results from different case studies:
# 1) Breast Cancer metrics (tabular data - RF, LogReg
bc_metrics = BASE / "case_studies/ds_lifecycle/outputs/metrics_breast_cancer.json"
if bc_metrics.exists():
    print("\nBreast Cancer metrics:")
    pprint.pp(json.loads(bc_metrics.read_text()))
    # show img
    for img in ["confusion_logreg.png","confusion_rf.png","roc_logreg.png","roc_rf.png"]:
        p = bc_metrics.parent / img
        if p.exists():
            display(Image(filename=str(p)))
else:
    print("\n[Note] Breast Cancer metrics not found.")

# 2) NLP metrics (text - LogReg & Naive Bayes)
nlp_metrics = BASE / "case_studies/nlp_sentiment/outputs/metrics_20news.json"
if nlp_metrics.exists():
    print("\nNLP metrics:")
    pprint.pp(json.loads(nlp_metrics.read_text()))
    for img in ["confusion_logreg.png","confusion_naive_bayes.png","confusion_logreg.png","confusion_naive_bayes.png"]:
        p = nlp_metrics.parent / img
        if p.exists():
            display(Image(filename=str(p)))
else:
    print("\n[Note] NLP metrics not found.")

# 3) Digits (images - MLP neuralna mrea)  to run image classification:
print("\nDigits MLP: run `python scripts/run_digits_mlp.py` (results appear in outputs)")
```
